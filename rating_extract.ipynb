```python
import numpy
import json
```

```python
fp =  open("D:/maimai_teisu_analyzer_project/乐谱.csv", 'r', encoding='utf-8')
data = fp.readlines()
del data[0]
level_dict = {}
```

```python
#写入歌曲定数
for song in data:
    song.strip('\n')
    temp = song.split(',')

    #若没有该曲目则新建
    #曲目名称->谱面种类->谱面颜色->定数
    if len(temp) >= 8:
        name = ''.join(temp[:-7])
        temp = [name] + temp[-7:]
    if level_dict.get(temp[0],0) == 0:
        level_dict[temp[0]] = {}
        level_dict[temp[0]]['SD'] = {}
        level_dict[temp[0]]['DX'] = {}

        for diff in ['Basic','Advanced','Expert','Master','Re:MASTER']:
            level_dict[temp[0]]['SD'][diff] = None
            level_dict[temp[0]]['DX'][diff] = None

    #录入当前曲目难度
    print(temp[0])
    level_dict[temp[0]][temp[1]][temp[2]] = float(temp[4])
    #try:
        #level_dict[temp[0]][temp[1]][temp[2]] = float(temp[4])
    #except:
        #pass
```

```python
with open('D:/maimai_teisu_analyzer_project/rating.json','w',encoding='utf-8') as rating:
    json.dump(level_dict, rating, indent=4)
```
